```python
# IMPORT ALL THE LIBRARIES USED IN THE NOTEBOOK
import matplotlib.pyplot as plt
import numpy as np
from waffles.utils.denoising.tv1ddenoise import Denoise
```

```python
cycles = 2 # how many sine cycles
resolution = 2000 # how many datapoints to generate

length = np.pi * 2 * cycles
my_wave = np.sin(np.arange(0, length, length / resolution))
noise = np.random.normal(0,0.5,resolution)
my_wave_noise = my_wave+noise

denoiser = Denoise()
```

```python
plt.plot(my_wave_noise)
plt.plot(my_wave)
plt.plot(denoiser.apply_denoise(my_wave_noise,1.5))
```
